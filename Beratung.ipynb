```python
import pandas as pd

import matplotlib.pyplot as plt
```

```python
data = pd.read_excel("beratung_ist.xlsx")
```

```python
data["And_Sprache"] = data['Andere…2'] + data['Andere…3'] 
```

```python
data = data.drop(columns=['Andere…2',
       'Andere…3'])
```

```python
data.rename(columns={'Afghan-\nistan':"Afghanistan", 'Uk\nraine':"Ukraine"}, inplace=True)
```

```python

plt.style.use('fivethirtyeight')
plt.figure(figsize=(12, 7), dpi=120)

slices =[data[column].sum() for column in data.columns[7:15]]
labels = data.columns[7:15]
colors = ["#9EA1D4",  "#FFF8E1", "#F5EA5A","#FF8787", "#FFB100", "#EB455F", "#85586F", "#86C8BC"]
plt.title("Beratung in 2022",)

plt.pie(slices, colors=colors,shadow=True, labels=labels, autopct="%.2f")

plt.axis('equal') 
plt.legend()
plt.savefig("plot_pie.png")
plt.show()
```
